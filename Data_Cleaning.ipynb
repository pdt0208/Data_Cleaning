```python
df1['Actual gross'] = df1['Actual gross'].replace({
    '$229,100,000[b]': '$229,100,000',
    '$167,700,000[e]': '$167,700,000'
})
```

```python
cols = ['Actual gross', 'Adjusted gross (in 2022 dollars)', 'Average gross']

df1[cols] = (
    df1[cols]
    .apply(lambda x: x.str.replace('$', '', regex=False)
                     .str.replace(',', '', regex=False))
)
```
